1. List — Find the Largest Number
You have a list of numbers: [34, 12, 89, 5, 67, 23]
Without using the max() function, write a program to find the largest number using a loop.

```python
List_of_Numbers = [34, 12, 89, 5, 67, 23]
largest = List_of_Numbers[0]
for i in List_of_Numbers:
	if i > largest:
		largest = i
print(largest)
```

2. Dictionary — Student Grades
Create a dictionary with 5 students and their marks. Print only the names of students who scored more than 70.
{"Amit": 85, "Priya": 60, "Rahul": 72, "Sara": 45, "John": 91}

```python
Students = {
    "Amit": 85,
    "Priya": 60,
    "Rahul": 72,
    "Sara": 45,
    "John": 91
}
print("Students who scored more than 70:")
for name, marks in Students.items():
    if marks > 70:
        print(name)
```

5. List + Dictionary — Inventory Counter
You have a list of sold items:
pythonsales = ["apple", "banana", "apple", "orange", "banana", "apple", "mango", "banana"]
Write a program that counts how many times each item was sold and stores it in a dictionary. Then print the most sold item.

```python
# List of sold items
sales = ["apple", "banana", "apple", "orange", "banana", "apple", "mango", "banana"]

# Count each item using a dictionary
inventory = {}
for item in sales:
    if item in inventory:
        inventory[item] += 1
    else:
        inventory[item] = 1

print(f"Sales Count: {inventory}")

# Find the HIGHEST count first
highest_count = max(inventory.values())

# Find ALL items that match the highest count
most_sold = [item for item, count in inventory.items() if count == highest_count]

print(f"Most Sold Item(s): {most_sold} ({highest_count} times each)")
```

6. All Four — Student Report Card
Create a program that:

Stores student names in a tuple (so names can't be changed)
Stores their subject marks in a dictionary where each student has a list of 3 subject marks
Uses a set to track which students scored above 75 in all 3 subjects
Prints the set of top performers

python students = ("Amit", "Priya", "Rahul", "Sara")
marks = {
    "Amit":  [80, 90, 70],
    "Priya": [60, 55, 78],
    "Rahul": [88, 92, 85],
    "Sara":  [40, 60, 55]
}

```python
students = ("Amit", "Priya", "Rahul", "Sara") #stored in tuple so name can not be change

marks = {
    "Amit" : [80, 90, 70],
    "Priya" : [60, 55, 68],
    "Rahul" : [88, 92, 95],
    "Sara" : [40, 60, 55]
}

top_performers = set()

for student in students:
    subject_marks = marks[student]
    if all(mark > 75 for mark in subject_marks):  # checks every mark
        top_performers.add(student)

print(f"All Students   : {students}")
print(f"Top Performers : {top_performers}")
```
